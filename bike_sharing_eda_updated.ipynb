## Standardize Attribute Names

```python
hour_df.rename(columns={'instant':'rec_id',
                      'dteday':'datetime',
                      'holiday':'is_holiday',
                      'workingday':'is_workingday',
                      'weathersit':'weather_condition',
                      'hum':'humidity',
                      'mnth':'month',
                      'cnt':'total_count',
                      'hr':'hour',
                      'yr':'year'},inplace=True)
```

## Typecast Attributes 

```python
# date time conversion
hour_df['datetime'] = pd.to_datetime(hour_df.datetime)

# categorical variables
hour_df['season'] = hour_df.season.astype('category')
hour_df['is_holiday'] = hour_df.is_holiday.astype('category')
hour_df['weekday'] = hour_df.weekday.astype('category')
hour_df['weather_condition'] = hour_df.weather_condition.astype('category')
hour_df['is_workingday'] = hour_df.is_workingday.astype('category')
hour_df['month'] = hour_df.month.astype('category')
hour_df['year'] = hour_df.year.astype('category')
hour_df['hour'] = hour_df.hour.astype('category')
```

## Interpreting Correlation and Noise in Regression

### Strong Correlation:
- A **strong correlation** (either positive or negative) between a feature and the target variable indicates that the feature has a meaningful relationship with the target.
- Features with strong correlations provide valuable information to the model, improving its ability to make accurate predictions.
- **Low noise**: Strongly correlated features are less likely to introduce noise, as they are directly related to the target variable and help in building a predictive model.

### Weak Correlation:
- A **weak correlation** (near zero) suggests that there is little or no linear relationship between the feature and the target variable.
- Weakly correlated features may be considered **noisy** since they do not contribute significant information to predicting the target.
- **High noise**: Features with weak correlations are more likely to introduce noise, which can lead to overfitting and reduced model performance.

### Noise in the Context of Features:
- Features with **weak or near-zero correlation** are more likely to be noise, as they do not have a strong linear relationship with the target variable.
- **Removing or transforming** such features can help reduce noise and improve the overall performance of the regression model.

By analyzing correlation, we can identify which features are contributing meaningful signal and which ones may be adding unnecessary noise to the model.


```python
# Select only the specified features
selected_features = ["temp", "atemp", "humidity", "windspeed", "casual", "registered", "total_count"]

# Subset the DataFrame for the selected features
hour_df_selected = hour_df[selected_features]

# Calculate correlations with `total_count`
continuous_corr = hour_df_selected.drop(columns=["total_count"]).corrwith(hour_df_selected["total_count"])

# Create a DataFrame for plotting
correlation_results = pd.DataFrame({
    'Feature': continuous_corr.index,
    'Correlation': continuous_corr.values
}).sort_values(by='Correlation', ascending=False)

# Plot
plt.figure(figsize=(8, 4))  # Smaller dimensions
plt.bar(correlation_results['Feature'], correlation_results['Correlation'], color='skyblue', width=0.6)
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)
plt.title('Correlation of Selected Features with Total Count', fontsize=10)  # Smaller title
plt.xlabel('Features', fontsize=8)  # Smaller x-axis label
plt.ylabel('Correlation', fontsize=8)  # Smaller y-axis label
plt.xticks(rotation=45, ha='right', fontsize=8)  # Smaller x-axis ticks
plt.yticks(fontsize=8)  # Smaller y-axis ticks
plt.tight_layout()
plt.show()
```
